```python
import os
import numpy as np
import matplotlib.pyplot as plt

input_file = "Outputs/data/MysteryData15020.txt"
data = np.loadtxt(input_file)
y = data[:]
x = np.arange(len(y))
plt.scatter(x, y, marker='o', linestyle='-', color='b')
plt.title("Mystery Data Plot")
plt.xlabel("Index")
plt.ylabel("Value")

```
